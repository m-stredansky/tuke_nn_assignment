```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-1, 1))
X_scaled1 = scaler.fit_transform(X_data2)

sns.histplot(X_scaled1, kde=True, color='blue', bins=50)
plt.title('Distribution of attributes after MinMax normalization')
plt.xlabel('Attributes')
plt.ylabel('Frequency')
plt.grid(True)
plt.legend().remove()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled2 = scaler.fit_transform(X_data2)

sns.histplot(X_scaled2, kde=True, color='blue', bins=50)
plt.title('Distribution of attributes after Standard normalization')
plt.xlabel('Attributes')
plt.ylabel('Frequency')
plt.grid(True)
plt.legend().remove()
plt.show()
```
