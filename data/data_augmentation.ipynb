```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel(r"C:\Users\firda\OneDrive\Študijný materiál\VŠ\2024 LS 2.rocnik\NeuronoveSiete\Zadanie\moje data\15min_dataset_pre_NN.xlsx", sheet_name="Sheet1")

X = data.drop(columns=["Rozsev 0-5"])
y = data["Rozsev 0-5"]
```

```python
def add_gaussian_noise_to_data(data, target, mean=0, std=0.1):
    """
    Funkcia pridáva Gaussovský šum k numerickým dátam aj cieľovým premenným.

    Parameters:
    - data: pandas.DataFrame - Vstupné numerické dáta.
    - target: pandas.Series - Cieľové premenné.
    - mean: float - Priemer Gaussovského šumu (predvolená hodnota: 0).
    - std: float - Štandardná odchýlka Gaussovského šumu (predvolená hodnota: 0.1).

    Returns:
    - noisy_data: pandas.DataFrame - Numerické dáta s pridaným Gaussovským šumom.
    - noisy_target: pandas.Series - Cieľové premenné s pridaným Gaussovským šumom.
    """
    noise_data = np.random.normal(mean, std, size=data.shape)
    noisy_data = data + noise_data
    
    noise_target = np.random.normal(mean, std, size=target.shape)
    noisy_target = target + noise_target
    
    return noisy_data, noisy_target
```

```python
X_noisy, y_noisy = add_gaussian_noise_to_data(X, y)
X_noisy
```

```python
y_noisy
```

```python
noisy_data = pd.concat([X_noisy, y_noisy], axis=1)
```

```python
# Pripojenie zašumených dát k pôvodným dátam
combined_data = pd.concat([data, noisy_data], ignore_index=True)
# Uloženie kombinovaných dát do nového súboru
combined_data.to_excel("kombinovane_data.xlsx", index=False)
```
