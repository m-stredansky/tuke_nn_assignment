```python
import pandas as pd
import numpy as np
```

```python
rozsevy_path = 'Rozsevy_Aglo_2016_2019.xlsx'
```

```python
rozsevy_sp1_df = pd.read_excel(rozsevy_path, index_col=0, sheet_name='SP1')
```

```python
rozsevy_sp1_df['Sinter.SP1.Rozsev.Under5mm.%'] = rozsevy_sp1_df['Sinter.SP1.Rozsev.Under5mm.%'].astype(str)
```

```python
sp1_df = rozsevy_sp1_df[~rozsevy_sp1_df['Sinter.SP1.Rozsev.Under5mm.%'].str.contains('Not Enough Values For Calculation')]

sp1_df = sp1_df.drop(columns=['Unnamed: 2'])
```

```python
sp1_df['Unnamed: 3'] = sp1_df['Unnamed: 3'] - pd.Timedelta(minutes=1)
```

# SP1 - 2016 - 2017 - 2018 - 2019

```python
interval_sp1_path_2016 = 'PI_DataLink_Aglo_digital twins-5min-SP1-2016-30minute.xlsm'
interval_sp1_path_2017 = 'PI_DataLink_Aglo_digital twins-5min-SP1-2017-30minute.xlsm'
interval_sp1_path_2018 = 'PI_DataLink_Aglo_digital twins-5min-SP1-2018-30minute.xlsm'
interval_sp1_path_2019 = 'PI_DataLink_Aglo_digital twins-5min-SP1-2019-30minute.xlsm'
```

```python
sp1_interval_df_2016 = pd.read_excel(interval_sp1_path_2016)
sp1_interval_df_2017 = pd.read_excel(interval_sp1_path_2017)
sp1_interval_df_2018 = pd.read_excel(interval_sp1_path_2018)
sp1_interval_df_2019 = pd.read_excel(interval_sp1_path_2019)
```

```python
sp1_interval_df_2016 = sp1_interval_df_2016.iloc[2:, 2:]
new_header1 = sp1_interval_df_2016.iloc[0]
sp1_interval_df_2016 = sp1_interval_df_2016[1:]
sp1_interval_df_2016.columns = new_header1
new_header1[0] = 'Datum od'
new_header1[1] = 'Datum do'
sp1_interval_df_2016 = sp1_interval_df_2016.reset_index(drop=True)

sp1_interval_df_2017 = sp1_interval_df_2017.iloc[2:, 2:]
new_header2 = sp1_interval_df_2017.iloc[0]
sp1_interval_df_2017 = sp1_interval_df_2017[1:]
sp1_interval_df_2017.columns = new_header2
new_header2[0] = 'Datum od'
new_header2[1] = 'Datum do'
sp1_interval_df_2017 = sp1_interval_df_2017.reset_index(drop=True)

sp1_interval_df_2018 = sp1_interval_df_2018.iloc[2:, 2:]
new_header3 = sp1_interval_df_2018.iloc[0]
sp1_interval_df_2018 = sp1_interval_df_2018[1:]
sp1_interval_df_2018.columns = new_header3
new_header3[0] = 'Datum od'
new_header3[1] = 'Datum do'
sp1_interval_df_2018 = sp1_interval_df_2018.reset_index(drop=True)

sp1_interval_df_2019 = sp1_interval_df_2019.iloc[2:, 2:]
new_header4 = sp1_interval_df_2019.iloc[0]
sp1_interval_df_2019 = sp1_interval_df_2019[1:]
sp1_interval_df_2019.columns = new_header4
new_header4[0] = 'Datum od'
new_header4[1] = 'Datum do'
sp1_interval_df_2019 = sp1_interval_df_2019.reset_index(drop=True)
```

```python
sp1_interval_all = pd.concat([sp1_interval_df_2016, sp1_interval_df_2017, sp1_interval_df_2018, sp1_interval_df_2019])
sp1_interval_all
```

```python
sp1_interval_all['Datum od'] = pd.to_datetime(sp1_interval_all['Datum od'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
sp1_merged_df = sp1_interval_all.merge(sp1_df, how='left', left_on='Datum od', right_on='Unnamed: 3')
sp1_merged_df.rename(columns={'Sinter.SP1.Rozsev.Under5mm.%': 'Sinter.Rozsev.Under5mm.%'}, inplace=True)

sp1_merged_df.reset_index(drop=True, inplace=True)
```

```python
nn_input_df = sp1_merged_df[sp1_merged_df['Sinter.SP1.Rozsev.Under5mm.%_y'].notna()]

nn_input_df
```

```python
nn_input_df = nn_input_df.loc[:, nn_input_df.columns.notna()]
nn_input_df.reset_index(drop=True, inplace=True)
```

```python
stlpce_na_odstranenie = [
    nn_input_df.columns[-4], nn_input_df.columns[-3], nn_input_df.columns[-2]
]

data = nn_input_df
data = nn_input_df.drop(columns=stlpce_na_odstranenie)
data = data.drop('Datum od', axis=1)
data = data.drop('Datum do', axis=1)
```

```python
data = data.drop('Sinter.SP1.C.Content.%', axis=1)
data = data.drop('Sinter.SP1.FeO.Content.%', axis=1)
data = data.drop('Sinter.SP1.QAGL.SinterProduction.Actual.Weight.t/hr', axis=1)
data = data.drop('Sinter.SP1.QZM.MixFlow.Actual.Weight.t/hr', axis=1)
data = data.drop('Sinter.SP1.Rozsev.Under5mm.%_x', axis=1)
data = data.drop('Sinter.SP1.SinterProduction.Weight.8Hours.SUM', axis=1)
data = data.drop('Sinter.TEX1.OTAC.Fan1.RPM', axis=1)
data = data.drop('Sinter.TEX2.OTAC.Fan2.RPM', axis=1)
data = data.drop('Sinter.EO1.T_PRED.Comb.Gas.Input.Temp.C', axis=1)
data = data.drop('Sinter.MB1.MS61.MixScales.Flow.t/hr', axis=1)
data = data.drop('Sinter.MB1.MS62.MixScales.Flow.t/hr', axis=1)
data = data.drop('Sinter.MB1.P1N.MixingDrumNo1.Permeability.kPa', axis=1)
data = data.drop('Sinter.MB1.MS21.CokeScales.Flow.t/hr', axis=1)
data = data.drop('Sinter.MB1.T1.MixingDrumNo1.Temperature.C', axis=1)
```

```python
data = data.drop('Sinter.LF1.Input.EO.Comb.Gas.Volume', axis=1)
data = data.drop('Sinter.LF1.AMS.Flow', axis=1)
data = data.drop('Sinter.LF1.Input.EO.H2O.Content', axis=1)
data = data.drop('Sinter.LF1.Input.EO.Comb.Gas.O2.Content', axis=1)
data = data.drop('Sinter.LF1.AMS.O2.Content', axis=1)
data = data.drop('Sinter.LF1.TEX1.Speed', axis=1)
data = data.drop('Sinter.LF1.TEX2.Speed', axis=1)
data = data.drop('Sinter.SP.C.Content.Average.%', axis=1)
```

```python
data = data.replace('[-11059] No Good Data For Calculation', np.nan)
data = data.dropna()
data.reset_index(drop=True, inplace=True)
data = data.rename(columns={'Sinter.SP1.Rozsev.Under5mm.%_y': 'Rozsev5mm'})
```
