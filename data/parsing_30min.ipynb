```python
import pandas as pd
```

```python
vstupny_datum = pd.read_excel('RozsAgloSP4-2018-vystup.xlsx', usecols=[0,5,6,7])
vstupny_datum['Dátum a čas'] = pd.to_datetime(vstupny_datum['Dátum a čas'])



vstupny_datum
```

```python
# Define the target times
time1 = pd.Timestamp('2024-04-13 23:00').time()
time2 = pd.Timestamp('2024-04-14 11:00').time()

# Iterate through each row and adjust the time while preserving the date
for index, row in vstupny_datum.iterrows():
    time = row['Dátum a čas'].time()
    if time < time2:
        new_datetime = row['Dátum a čas'] - pd.Timedelta(days=1)
        new_datetime = pd.Timestamp.combine(new_datetime.date(), time1)
    elif time2 <= time < time1:
        new_datetime = pd.Timestamp.combine(row['Dátum a čas'].date(), time2)
    else:
        new_datetime = pd.Timestamp.combine(row['Dátum a čas'].date(), time1)
    vstupny_datum.at[index, 'Dátum a čas'] = new_datetime
    
vstupny_datum
```

```python
vystupny_df = pd.read_excel('PI_DataLink_Aglo_digital twins-5min-SP4-2018-30minute.xlsm')

# vystupny_df = vystupny_df.iloc[3:]
vystupny_df = vystupny_df.iloc[3:, 2:]
vystupny_df
```

```python
vstupny_datum['date'] = pd.to_datetime(vstupny_datum['Dátum a čas']).dt.date
vystupny_df['date'] = pd.to_datetime(vystupny_df['Unnamed: 2']).dt.date

datumy = pd.merge(vstupny_datum, vystupny_df, on='date' , how='inner')

datumy

```

```python
final = datumy[(datumy['Dátum a čas'] >= datumy['Unnamed: 2']) & (datumy['Dátum a čas'] < datumy['Unnamed: 3'])]

final_20 = final.iloc[:,:67]
final_20
```
